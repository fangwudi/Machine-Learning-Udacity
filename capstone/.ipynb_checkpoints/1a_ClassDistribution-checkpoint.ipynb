```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
driver_imgs_list = pd.read_csv(basedir+'/input/driver_imgs_list.csv')
```

```python
driver_imgs_list.head()
```

```python
len(driver_imgs_list)
```

```python
subject_list = list(driver_imgs_list['subject'].unique())
```

```python
len(subject_list)
```

```python
print(subject_list)
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x="classname", data=driver_imgs_list, color='royalblue')
plt.ylabel('Count', fontsize=18)
plt.xlabel('Class', fontsize=18)
plt.ylim(0,3000)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title("Frequency of 10 classes", fontsize=22)
plt.show()
```

```python
fig, axs = plt.subplots(ncols=6, nrows=5, figsize=(30,30))

for i in range(5):
    for j in range(6):
        n = i*6+j
        if (n+1) > len(subject_list):break
        subject = subject_list[n]
        ax=axs[i][j]
        sns.countplot(x="classname", data=driver_imgs_list[driver_imgs_list['subject']==subject],\
                      color='royalblue', ax=ax)
        ax.set_ylabel('Count', fontsize=10)
        ax.set_xlabel('Class', fontsize=10)
        ax.set_title(subject, fontsize=20)
    
```
